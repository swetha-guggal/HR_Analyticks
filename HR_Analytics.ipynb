# Data cleaning


```python
#Renaming column names
#Define a dictionary to map old column names to new column names
new_column_names = {
    'Age': 'EmployeeAge',
    'DistanceFromHome': 'DistanceFromOffice',
    'MonthlyIncome': 'MonthlySalary',
    'DailyRate': 'DailyPayment',
    'Department': 'Dept',
    'Education': 'EducationLevel',
    'EducationField': 'FieldOfStudy',
    'EmployeeNumber': 'EmployeeID',
    'Gender': 'Sex',
    'JobInvolvement': 'InvolvementLevel',
    'JobLevel': 'PositionLevel',
    'JobRole': 'RoleAtWork',
    'JobSatisfaction': 'JobSatisfactionLevel',
    'MaritalStatus': 'MaritalStatus',
    'MonthlyRate': 'MonthlyRate',
    'NumCompaniesWorked': 'CompaniesWorked',
    'Over18': 'OverEighteen',
    'OverTime': 'Overtime',
    'PercentSalaryHike': 'SalaryHikePercentage',
    'PerformanceRating': 'PerformanceRating',
    'RelationshipSatisfaction': 'RelationshipSatisfactionLevel',
    'StandardHours': 'StandardHours',
    'StockOptionLevel': 'StockOptionsLevel',
    'TotalWorkingYears': 'TotalYearsWorked',
    'TrainingTimesLastYear': 'TrainingLastYear',
    'WorkLifeBalance': 'LifeBalance',
    'YearsAtCompany': 'YearsAtCurrentCompany',
    'YearsInCurrentRole': 'YearsInCurrentPosition',
    'YearsSinceLastPromotion': 'YearsSinceLastPromotion',
    'YearsWithCurrManager': 'YearsWithCurrentManager'
}

# Rename the columns in your DataFrame
data.rename(columns=new_column_names, inplace=True)

# Display the DataFrame with renamed columns
data.head()
```


  Redundant columns are those that do not provide unique or valuable information for our analysis.

```python
#Removing null values
#Remove rows with any NaN values from the entire DataFrame
cleaned_df = data.dropna()

# Display the cleaned DataFrame
cleaned_df.head()
```

```python
#Dropping Duplicates:
cleaned_df = cleaned_df.drop_duplicates()
cleaned_df
```

#Data Visualization:

#correlation map
#Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns you want to include in the correlation map
columns_for_correlation = ['EmployeeAge', 'MonthlySalary', 'TotalYearsWorked', 'YearsAtCurrentCompany', 'YearsInCurrentPosition', 'YearsWithCurrentManager']

# Calculate the correlation matrix
correlation_matrix = cleaned_df[columns_for_correlation].corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Map")
plt.show()
```

Summary of the "HR Analytics" Project:

The "HR Analytics" project involves the analysis of human resources data to gain insights and make data-driven decisions related to employee attrition, job satisfaction, and various HR-related factors. The project can be divided into two main phases: data cleaning and data visualization.

Data Cleaning:

Redundant Columns: Remove redundant columns that do not provide valuable information for analysis.
Renaming Columns: Rename columns to make them more descriptive and user-friendly.
Duplicate Rows: Identify and remove duplicate rows if they exist.

Correlation Map: Create a correlation map to visualize the relationships between numerical variables.
Overtime Analysis: Visualize the impact of overtime on various HR-related factors.
Marital Status Analysis: Explore how marital status relates to attrition and other aspects.
Job Role Analysis: Analyze job roles and their impact on attrition and job satisfaction.
Gender Analysis: Investigate gender-related trends in the workplace.
Education Field Analysis: Explore how the field of education impacts HR-related factors.
Department Analysis: Study the departments in the organization and their effect on attrition.
Business Travel Analysis: Examine the relationship between business travel and HR aspects.
Age and Overtime Analysis: Understand the connection between age and overtime work.
Total Working Years: Visualize the distribution of total working years among employees.
Education Level: Investigate the education level of the workforce and its impact.
Number of Companies Worked: Analyze how the number of companies an employee worked for relates to other factors.
Distance from Home: Study the distance of the employee's home from the workplace.
The project's primary goal is to gain insights from the data that can help the HR department make informed decisions regarding employee retention, job satisfaction, and other HR-related aspects. By visualizing and analyzing the data.
